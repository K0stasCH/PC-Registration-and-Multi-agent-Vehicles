```python
def filterResults(errors, transThres=5, AngleError=10):
  errors = errors[errors[:,0]<transThres]
  errors = errors[errors[:,1]<AngleError]
  errors = errors[errors[:,2]<AngleError]
  errors = errors[errors[:,3]<AngleError]
  return errors
```

```python
def eval(v1, v2, timeRange, cluster_selection_epsilon, min_cluster_size, ICP_thres, ICP_classes):
  errors_before = []
  errors_after = []

  for t in range(timeRange[0],timeRange[1]):

    Scene1, masks1 = v1.generateScene(model, t)
    Scene2, masks2 = v2.generateScene(model, t)

    g1 = Scene1.generateGraph(_min_cluster_size=min_cluster_size,
                                _cluster_selection_epsilon=cluster_selection_epsilon,
                                classNames=['person', 'vehicle', 'pole'])
    g2 = Scene2.generateGraph(_min_cluster_size=min_cluster_size,
                                _cluster_selection_epsilon=cluster_selection_epsilon,
                                classNames=['person', 'vehicle', 'pole'])

    affM, matchMatrix = matchGraph(g1, g2)
    TransMatrix_GT = get_GT_tranfMatrix_V1_2_V2(v1, v2, t)
    TransMatrix = getTransforamtionMatrix_from_MatchingMatrix(g1, g2, matchMatrix, 3)
    regResult = performICP(Scene1, Scene2, TransMatrix, threshold=ICP_thres, classes=ICP_classes)
    T_final = copy.deepcopy(regResult.transformation)

    e_before = printErrors(TransMatrix_GT, TransMatrix, 'degrees')
    e_after = printErrors(TransMatrix_GT, T_final, 'degrees')
    errors_before.append(e_before)
    errors_after.append(e_after)

  errors_before = np.array(errors_before)
  errors_after = np.array(errors_after)

  errors_before[:,1:4] = (errors_before[:,1:4]+180)%360-180
  errors_after[:,1:4] = (errors_after[:,1:4]+180)%360-180

  return np.abs(errors_before), np.abs(errors_after)
```

```python
classesICP = ['building', 'pole', 'traffic light', 'person']
examples = [[1,4,[0,100]],
            [2,1,[80,100]],
            [2,4,[80,100]],
            [2,5,[80,100]],
            [4,5,[0,100]]]

cluster_selection_epsilon = 0.5
min_cluster_size = 10

ICP_t = 1
errorsBefore = np.empty([0,4])
errorsAfter = np.empty([0,4])
for i, e in enumerate(examples):
  print(f'----------------------{i}---------------------------------')
  v1 = Vehicle(v2x_sim,e[0])
  v2 = Vehicle(v2x_sim,e[1])
  before, after = eval(v1, v2, e[2], cluster_selection_epsilon, min_cluster_size, ICP_t, classesICP)
  errorsAfter = np.vstack([errorsAfter, after])
  errorsBefore = np.vstack([errorsBefore, before])
```

```python
rA = filterResults(errorsAfter)
rB = filterResults(errorsBefore)
```
